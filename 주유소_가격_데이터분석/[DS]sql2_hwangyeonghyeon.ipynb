```python
import csv
import mysql.connector

conn = mysql.connector.connect(
    host='database-1.cn4vthwm7kuj.us-east-2.rds.amazonaws.com',
    user='oneday',
    password='1234',
    database='oneday'
)

csv_file_path = 'gas_station_data.csv'

query = '''
SELECT GAS_STATION.id, GAS_BRAND.name AS brand, GAS_STATION.name, GAS_STATION.city, GAS_STATION.gu, GAS_STATION.address, 
       GAS_STATION.gasoline, GAS_STATION.diesel, GAS_STATION.self, GAS_STATION.car_wash, GAS_STATION.charging_station, 
       GAS_STATION.car_maintenance, GAS_STATION.convenience_store, GAS_STATION.24_hours, GAS_STATION.lat, GAS_STATION.lng
FROM GAS_STATION
JOIN GAS_BRAND ON GAS_STATION.brand = GAS_BRAND.id
ORDER BY GAS_STATION.id;
'''

cursor = conn.cursor()
cursor.execute(query)
results = cursor.fetchall()

# CSV 파일 저장
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['id', 'brand', 'name', 'city', 'gu', 'address', 'gasoline', 'diesel', 'self', 'car_wash',
                     'charging_station', 'car_maintenance', 'convenience_store', '24_hours', 'lat', 'lng'])
    writer.writerows(results)

cursor.close()
conn.close()

print('데이터가 CSV 파일로 저장되었습니다.')
```

위도, 경도 정보를 이용하여 미왕빌딩에서 1킬로 이내에 위치한 주유소 정보를 검색하세요. 

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 미왕빌딩으로 부터의 거리 (km)

- 쿼리, 결과

```python
import mysql.connector

conn = mysql.connector.connect(
    host='database-1.cn4vthwm7kuj.us-east-2.rds.amazonaws.com',
    user='oneday',
    password='1234',
    database='oneday'
)


# 미왕빌딩 위도와 경도 설정
miwang_building_lat = 37.495435686811
miwang_building_lng = 127.02915553846

query = """
SELECT GAS_STATION.id, GAS_BRAND.name AS brand, GAS_STATION.name, address,
    ROUND(
        ST_Distance_Sphere(
            POINT(%s, %s),
            POINT(lng, lat)
        ) / 1000, 2
    ) AS distance
FROM GAS_STATION
JOIN GAS_BRAND ON GAS_STATION.brand = GAS_BRAND.id
WHERE
    ST_Distance_Sphere(
        POINT(%s, %s),
        POINT(lng, lat)
    ) < 1000
ORDER BY distance;
"""
params = (miwang_building_lng, miwang_building_lat, miwang_building_lng, miwang_building_lat)


cursor = conn.cursor()
cursor.execute(query, params)

# 결과 출력
for row in cursor.fetchall():
    gas_station_id, brand, name, address, distance = row
    print(f"주유소 아이디: {gas_station_id}")
    print(f"주유소 브랜드명: {brand}")
    print(f"주유소 매장명: {name}")
    print(f"주소: {address}")
    print(f"미왕빌딩으로부터의 거리 (km): {distance}")
    print('---------------------------------------------')

cursor.close()
conn.close()
```

```python
query = """
SELECT gu, brand_name, average_price
FROM (
  SELECT gu, GAS_BRAND.name AS brand_name, AVG(gasoline) AS average_price,
         ROW_NUMBER() OVER (PARTITION BY gu ORDER BY AVG(gasoline)) AS rn
  FROM GAS_STATION
  JOIN GAS_BRAND ON GAS_STATION.brand = GAS_BRAND.id
  GROUP BY gu, brand_name
) AS subquery
WHERE rn = 1
ORDER BY average_price ASC;

"""

cursor.execute(query)

rows = cursor.fetchall()
current_gu = ""

for i, row in enumerate(rows, start=1):
    구_이름, 주유소_브랜드명, 휘발유_평균_가격 = row
    구_이름_with_순서 = f"{i}. {구_이름}"
    print(구_이름_with_순서)
    print("주유소 브랜드명:", 주유소_브랜드명)
    print("휘발유 평균 가격:", 휘발유_평균_가격)
    print("---------------------------------------------")
```
